## ETS Decomposition of Airline Passengers Dataset:

## Import libraries

```python
from statsmodels.tsa.seasonal import seasonal_decompose

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Read dataset

```python
airline = pd.read_csv('airline-passengers.csv',
                       index_col ='Month',
                       parse_dates = True)
```

## show dataset

```python
airline.head()
```

## Apply ETS

```python
result = seasonal_decompose(airline['Passengers'],
                            model ='multiplicative')
```

## Plot ETS

```python
result.plot()
```
