## Data Preprocessing
Do what you think you need such as:

```python
sns.boxplot(df)
plt.xticks(rotation=90)
plt.show()
```

```python
def remove_outliers(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR



    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

df= remove_outliers(df,['Location_Easting_OSGR','Location_Northing_OSGR'])
```

```python
sns.boxplot(df)
plt.xticks(rotation=90)
plt.show()
```

```python
df['Accident_Severity'] = df['Accident_Severity'].fillna(df['Accident_Severity'].mode()[0])
df['Number_of_Casualties'] = df['Number_of_Casualties'].fillna(df['Number_of_Casualties'].mode()[0])
df['Light_Conditions'] = df['Light_Conditions'].fillna(df['Light_Conditions'].mode()[0])
df['Urban_or_Rural_Area'] = df['Urban_or_Rural_Area'].fillna(df['Urban_or_Rural_Area'].mode()[0])
```

## Clustering
Apply K-Means clustering. Determine the optimal number of clusters and justify your choice.
* Find the `n_clusters` parameter using the elbow method.
* Train the model.

```python
inertia = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
  kmeans.fit(df_pca)
  inertia.append(kmeans.inertia_)

plt.plot(range(1,11), inertia, marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=0)

kmeans.fit(df_pca)

cluster_centers = kmeans.cluster_centers_

labels = kmeans.predict(df_pca)

print("Cluster Centers:", cluster_centers)
print("Cluster Labels:", labels)
```

## Evaluation
Evaluate the clustering result using appropriate metrics.


```python
silhouette_score(df_pca, labels)
```

## Plot the data points with their predicted cluster center

```python
plt.figure(figsize=(8, 6))
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=labels)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red',marker='X', label='Cluster Centers')
plt.title('K-means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
